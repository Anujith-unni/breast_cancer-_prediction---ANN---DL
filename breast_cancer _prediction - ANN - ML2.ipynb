```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('breast_cancer.csv')
data.head()
```

```python
data.info()
```

```python
### delete the columns
data = data.drop(['id','Unnamed: 32'],axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data['diagnosis'].value_counts()
# B - Benign(Negatie), M - Malignant(postibe)
```

```python
sns.countplot(x=data['diagnosis'])
plt.show()
```

```python
### label encoding  -  converting categorical string values to numerical values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
data.info()
```

```python
### check for multicolinearity
data.corr()
## radius_mean vs perimeter_mean,area_mean,concave points_mean,radius_worst
# perimeter_worst,area_worst
```

```python
c = data.corr()
filter1 = c['radius_mean'] >= 0.8
filter2 = c['radius_mean'] <= -0.8
## radius mean
col = c[filter1 | filter2].index
col = col[1:]
data = data.drop(col,axis=1)
```

```python
c = data.corr()
filter1 = c['texture_mean'] >= 0.8
filter2 = c['texture_mean'] <= -0.8
## radius mean
col = c[filter1 | filter2].index
col = col[1:]
data = data.drop(col,axis=1)
```

```python
data.head()
```

```python
sns.boxplot(x=data['radius_mean'])
plt.show()
```

```python
data['radius_mean'].describe()
```

```python
filter = data['radius_mean'] > 20
row_index = data[filter].index
data = data.drop(row_index,axis=0)
```

```python
sns.boxplot(x=data['radius_mean'])
plt.show()
```

```python
my_col = data.columns[1:]
plt.figure(figsize=(20,20))
for i in range(len(my_col)):
  plt.subplot(5,5,i+1)
  plt.title(my_col[i])
  sns.boxplot(x = data[my_col[i]])
plt.show()
```
